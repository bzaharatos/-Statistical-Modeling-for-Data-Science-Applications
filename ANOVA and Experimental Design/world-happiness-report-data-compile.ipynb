```python
import numpy as np
import pandas as pd
```

```python
wh2015 = pd.read_csv('Datasets/world-happiness-chunks/datasets_894_813759_2015.csv')
wh2015['Year'] = str(2015)

wh2016 = pd.read_csv('Datasets/world-happiness-chunks/datasets_894_813759_2016.csv')
wh2016['Year'] = str(2016)

wh2017 = pd.read_csv('Datasets/world-happiness-chunks/datasets_894_813759_2017.csv')
wh2017['Year'] = str(2017)

wh2018 = pd.read_csv('Datasets/world-happiness-chunks/datasets_894_813759_2018.csv')
wh2018['Year'] = str(2018)

wh2019 = pd.read_csv('Datasets/world-happiness-chunks/datasets_894_813759_2019.csv')
wh2019['Year'] = str(2019)
```

```python
rename_2015 = {'Happiness Rank':'Rank',
                 'Happiness Score':'Happiness_score',
                 'Economy (GDP per Capita)':'GDP_per_capita',
                 'Health (Life Expectancy)':'Health', 
                 'Trust (Government Corruption)':'Trust_in_government'}
wh2015 = wh2015[['Happiness Rank',
                 'Country',
                 'Happiness Score',
                 'Economy (GDP per Capita)',
                 'Family',
                 'Health (Life Expectancy)', 
                 'Freedom',
                 'Trust (Government Corruption)',
                 'Generosity',
                 'Year']].rename(columns = rename_2015)
```

```python
rename_2016 = {'Happiness Rank':'Rank',
                 'Happiness Score':'Happiness_score',
                 'Economy (GDP per Capita)':'GDP_per_capita',
                 'Health (Life Expectancy)':'Health', 
                 'Trust (Government Corruption)':'Trust_in_government'}
wh2016 = wh2016[['Happiness Rank',
                 'Country',
                 'Happiness Score',
                 'Economy (GDP per Capita)',
                 'Family',
                 'Health (Life Expectancy)', 
                 'Freedom',
                 'Trust (Government Corruption)',
                 'Generosity',
                 'Year']].rename(columns = rename_2016)
```

```python
rename_2017 = {'Happiness.Rank':'Rank',
                 'Happiness.Score':'Happiness_score',
                 'Economy..GDP.per.Capita.':'GDP_per_capita',
                 'Health..Life.Expectancy.':'Health', 
                 'Trust..Government.Corruption.':'Trust_in_government'}
wh2017 = wh2017[['Happiness.Rank',
                 'Country',
                 'Happiness.Score',
                 'Economy..GDP.per.Capita.',
                 'Family',
                 'Health..Life.Expectancy.', 
                 'Freedom',
                 'Trust..Government.Corruption.',
                 'Generosity',
                 'Year']].rename(columns = rename_2017)
```

```python
rename_2018 = {'Overall rank':'Rank',
                'Country or region':'Country',
                'Score':'Happiness_score',
                'GDP per capita':'GDP_per_capita',
                'Social support':'Family',
                'Healthy life expectancy':'Health', 
                'Freedom to make life choices':'Freedom',
                'Perceptions of corruption':'Trust_in_government'}
wh2018 = wh2018[['Overall rank',
                 'Country or region',
                 'Score',
                 'GDP per capita',
                 'Social support',
                 'Healthy life expectancy', 
                 'Freedom to make life choices',
                 'Perceptions of corruption',
                 'Generosity',
                 'Year']].rename(columns = rename_2018)
```

```python
rename_2019 = {'Overall rank':'Rank',
                'Country or region':'Country',
                'Score':'Happiness_score',
                'GDP per capita':'GDP_per_capita',
                'Social support':'Family',
                'Healthy life expectancy':'Health', 
                'Freedom to make life choices':'Freedom',
                'Perceptions of corruption':'Trust_in_government'}
wh2019 = wh2019[['Overall rank',
                 'Country or region',
                 'Score',
                 'GDP per capita',
                 'Social support',
                 'Healthy life expectancy', 
                 'Freedom to make life choices',
                 'Perceptions of corruption',
                 'Generosity',
                 'Year']].rename(columns = rename_2019)
```

```python
df = pd.concat([wh2015,wh2016,wh2017,wh2018,wh2019])
```
